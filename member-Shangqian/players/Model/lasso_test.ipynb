```python
 df2 = df.copy()[['Season', 'Pickup_month', 'Pickup_week', 'Pickup_day', 'Pickup_wday',
       'Weekend', 'Pickup_hour', 'Time', 'Pickup_minute', 'Borough', 'Zone',
       'Pickup_cell', 'Number_trips']]
```

```python
 df2 = df2.groupby(['Season', 'Pickup_month', 'Pickup_week', 'Pickup_day', 'Pickup_wday',
       'Weekend', 'Pickup_hour', 'Time', 'Pickup_minute', 'Borough', 'Zone',
       'Pickup_cell']).sum().reset_index()
```

```python
df2=df2.replace({'Season': {"Spring": 1, 'Summer': 2,'Autumn':3,'Winter':4}})
df2=df2.replace({'Season': {"Spring": 1, 'Summer': 2,'Autumn':3,'Winter':4}})
```

```python
 X = df2.iloc[:, 0:12]
y = df2.loc[:,"Number_trips"].astype('float')
```

```python
from sklearn.preprocessing import OneHotEncoder
o=OneHotEncoder()
```

```python
from sklearn import preprocessing
e = preprocessing.LabelEncoder()
```

```python
X["Pickup_month"]=e.fit_transform(X["Pickup_month"])
X["Pickup_wday"]=e.fit_transform(X["Pickup_wday"])

X["Weekend"]=e.fit_transform(X["Weekend"])

X["Time"]=e.fit_transform(X["Time"])
X["Borough"]=e.fit_transform(X["Borough"])
X["Zone"]=e.fit_transform(X["Zone"])
X["Pickup_cell"]=e.fit_transform(X["Pickup_cell"])
```

```python
train = o.fit_transform(X)
```

```python
from sklearn import linear_model
reg = linear_model.Ridge(alpha=.5)
```

```python
reg.fit(train,y)
```
